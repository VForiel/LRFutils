```python
import term
import time

N = 1000

a = term.progress(max = N)

for i in range(N):
    a(i+1)
    time.sleep(0.01)

a = term.progress(prefix = "The best bar", max = N)

for i in range(N):
    a(i+1)
    if i == N-300:
        a.stop()
        break
    time.sleep(0.01)

a = term.progress(prefix = "Waw, another incredible bar", max = N)

for i in range(N):
    a(i+1)
    time.sleep(0.01)
```
